각 사람(subject)의 활동 종류(event)에 따라 각 시점(timepoint) 별로 fmri(뇌활동 측정 기술)의 측정값 중 하나의 센서값을 정리한 
시계열 데이터

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
#subject와 event 구분 없이 timepoint 별로 반복적으로 관측된 값을 바탕으로 그림

#신뢰 구간과 추정 회귀선까지 그려짐

sns.relplot(x = 'timepoint', y = 'signal', kind='line',data=fmri)
```

```python
#신뢰구간 빼기

sns.relplot(x = 'timepoint', y = 'signal', kind='line',data=fmri, ci=None)
```

```python
sns.relplot(x='timepoint', y='signal', kind='line',data=fmri, hue ='region', style = 'event')
```
