```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 요일 별로 total_bill을 표현, 즉 범주형, 요일별로 정산 등에 유용
# 요일 별 확인이 좀 더 명확하더라 하는 장점

sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'boxen')
```

```python
#? 성별로 색을 구분해서 swarm 차트 형식의 catplot 그리기

sns.catplot(x = 'day', y = 'total_bill', data = tips, hue ='sex', kind='swarm')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue ='sex', kind='violin',
            split=True, inner='stick')
```

```python
# ? 타이타닉 데이터 활용해서 생존률 확인 가능한 차트를 그려보기
titanic = sns.load_dataset('titanic')
titanic.head(10)
```

```python
sns.catplot(x = 'age', y = 'alive', data = titanic, hue ='sex', kind='swarm')
```

```python
sns.catplot(x = 'sex', y = 'survived', data = titanic, hue='pclass', kind='bar', ci=None)
```

```python
sns.catplot(x = 'sex', y = 'survived', data = titanic, hue='pclass', kind='bar', ci=None)
```

```python
sns.countplot(x="class", hue="who", data=titanic)
```

```python
#하나의 Figure에 두 개의 chart(axes 두 개) 그리기

fig,ax = plt.subplots(1,2, figsize=(12,5))

titanic[['sex','survived']].groupby(['sex']).mean().plot.bar(ax=ax[0])

ax[0].set_title('Survived vs Sex')

sns.countplot(x='sex', hue = 'survived', data=titanic, ax=ax[1])
ax[1].set_title('Survived vs Dead')
```
